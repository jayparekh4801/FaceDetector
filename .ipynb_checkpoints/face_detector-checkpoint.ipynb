# importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('./baldFace.jpeg')
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
def convetToRBG(image) :
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def faceDetecor(cascader, image, scale_factor = 1.1) :
    image_copy = image.copy()
    faces = cascader.detectMultiScale(image, scaleFactor = scale_factor, minNeighbors = 5)
    for (x, y, w, h) in faces :
        cv2.rectangle(image_copy, (x, y), (x + w, y + h), (0, 255, 0), 2)

    return image_copy
```

```python
haar_cascade_face = cv2.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
plt.imshow(convetToRBG(faceDetecor(haar_cascade_face, image, 1.2)))
```
